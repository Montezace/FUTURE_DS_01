```python
import pandas as pd
import ipywidgets as widgets
from IPython.display import display, HTML
import matplotlib.pyplot as plt
import base64
from io import BytesIO
import datetime

# 1. Load the data (handles common Kaggle encoding issues)
try:
    df = pd.read_csv('data.csv', encoding='utf-8')
except UnicodeDecodeError:
    df = pd.read_csv('data.csv', encoding='unicode_escape')

print("Data loaded successfully! Please map your columns below:")

# 2. Setup Interactive Menus
date_col = widgets.Dropdown(options=['None'] + list(df.columns), description='Date:')
product_col = widgets.Dropdown(options=['None'] + list(df.columns), description='Product:')
region_col = widgets.Dropdown(options=['None'] + list(df.columns), description='Region/Cat:')
rev_col = widgets.Dropdown(options=['None'] + list(df.columns), description='Revenue:')
qty_col = widgets.Dropdown(options=['None'] + list(df.columns), description='Quantity:')

button = widgets.Button(description="Generate Client Report", button_style='success')
output = widgets.Output()

display(date_col, product_col, region_col, rev_col, qty_col, button, output)

# Helper function to convert plots to base64 for the HTML report
def plot_to_base64(plt_obj):
    img = BytesIO()
    plt_obj.savefig(img, format='png', bbox_inches='tight')
    img.seek(0)
    return base64.b64encode(img.getvalue()).decode()

# 3. Execution Pipeline
def generate_report(b):
    with output:
        output.clear_output()
        print("Processing data and generating report...")

        # Working copy
        data = df.copy()

        # Data Cleaning & Prep
        if date_col.value != 'None':
            data[date_col.value] = pd.to_datetime(data[date_col.value], errors='coerce')
            data = data.dropna(subset=[date_col.value])
            data['Month_Year'] = data[date_col.value].dt.to_period('M').astype(str)

        # Calculate Revenue if only Price and Quantity are provided
        if rev_col.value == 'None' and qty_col.value != 'None':
            # Assuming the user selects Price in the Revenue dropdown by mistake, or needs to calculate it
            print("Revenue column not selected. Calculating from metrics...")
            pass # In a real scenario, you might add a 'Price' dropdown to calculate rev = price * qty

        target_rev = rev_col.value if rev_col.value != 'None' else qty_col.value

        # Calculate KPIs
        total_revenue = data[target_rev].sum()

        # Trend Analysis Plot
        plt.figure(figsize=(10, 4))
        if date_col.value != 'None':
            trend = data.groupby('Month_Year')[target_rev].sum()
            trend.plot(kind='line', marker='o', color='#2E86C1', linewidth=2)
            plt.title('Revenue Trend Over Time')
            plt.ylabel('Revenue / Volume')
            plt.xticks(rotation=45)
            plt.grid(True, linestyle='--', alpha=0.6)
        trend_img = plot_to_base64(plt)
        plt.close()

        # Top Products Plot
        plt.figure(figsize=(10, 4))
        if product_col.value != 'None':
            top_products = data.groupby(product_col.value)[target_rev].sum().nlargest(5)
            top_products.sort_values().plot(kind='barh', color='#28B463')
            plt.title('Top 5 Products by Performance')
            plt.xlabel('Revenue / Volume')
        prod_img = plot_to_base64(plt)
        plt.close()

        # Top Region/Category Plot
        plt.figure(figsize=(10, 4))
        if region_col.value != 'None':
            top_regions = data.groupby(region_col.value)[target_rev].sum().nlargest(5)
            top_regions.plot(kind='bar', color='#884EA0')
            plt.title('Performance by Region / Category')
            plt.ylabel('Revenue / Volume')
            plt.xticks(rotation=45)
        reg_img = plot_to_base64(plt)
        plt.close()

        # 4. Generate HTML Report
        html_content = f"""
        <html>
        <head>
            <style>
                body {{ font-family: 'Segoe UI', Tahoma, Geneva, Verdana, sans-serif; margin: 40px; color: #333; }}
                h1 {{ color: #2C3E50; border-bottom: 2px solid #3498DB; padding-bottom: 10px; }}
                h2 {{ color: #2980B9; }}
                .kpi-box {{ background-color: #ECF0F1; padding: 20px; border-radius: 8px; text-align: center; font-size: 24px; font-weight: bold; margin-bottom: 30px; border-left: 5px solid #3498DB; }}
                .insight-box {{ background-color: #F9E79F; padding: 15px; border-radius: 5px; margin-top: 10px; font-style: italic; }}
                img {{ max-width: 100%; height: auto; border-radius: 8px; box-shadow: 0 4px 8px rgba(0,0,0,0.1); margin-bottom: 20px; }}
            </style>
        </head>
        <body>
            <h1>Business Sales Performance Analytics</h1>
            <p><strong>Generated on:</strong> {datetime.datetime.now().strftime('%Y-%m-%d')}</p>

            <div class="kpi-box">
                Total Performance Metric: {total_revenue:,.2f}
            </div>

            <h2>1. Revenue Trends Over Time</h2>
            <img src="data:image/png;base64,{trend_img}" alt="Trend Chart">
            <div class="insight-box">Insight: Look for seasonal spikes or dips in the chart above to optimize marketing spend during peak periods.</div>

            <h2>2. Top-Selling Products</h2>
            <img src="data:image/png;base64,{prod_img}" alt="Products Chart">
            <div class="insight-box">Insight: These top 5 products drive the majority of the metric. Ensure inventory levels for these items remain highly stocked.</div>

            <h2>3. High-Value Categories & Regional Performance</h2>
            <img src="data:image/png;base64,{reg_img}" alt="Region Chart">
            <div class="insight-box">Insight: Focus regional expansion and targeted ad campaigns on the highest-performing segments shown above.</div>

            <hr>
            <p style="text-align: center; color: #7F8C8D;">Task 1 Delivery | Future Interns Data Science & Analytics Track</p>
        </body>
        </html>
        """

        # Save file using the exact required naming format
        filename = "FUTURE_DS_01_Report.html"
        with open(filename, "w") as file:
            file.write(html_content)

        print(f"✅ Success! Report exported as '{filename}'.")
        print("You can now download this file from the Colab file browser and upload it to your GitHub repository.")

# Bind button
button.on_click(generate_report)
```
